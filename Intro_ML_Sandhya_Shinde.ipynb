**Predicting breast cancer using a machine learning model**

1. Problem statement:
To identify and create a machine learning model that predicts whether a patient has breast cancer. The goal is to find accuracy score of the breast cancer prediciton model with random seed set as 0.

2. Data:
The data set is placed at: https://github.com/sandhyashinde06/repo/blob/main/dataset_ML.csv

3. Evaluation:
This prediction aims at gaining accuracy score higher than 95.8% at predicting the patient has a malignant or benign tumor.

4. Data extraction and evaluation:
Since all the data is in the form of numbers we would not have to keep converting categorical data types to numeric types. Also as there are are no missing values, we do not worry about missing computed data.
In the heatmap we observe that the mean fractal dimension, texture error, smoothness error, symmetry error and fractal dimension errors influence the target variable and that there are more samples of benign than malignant tumors.

```python
import warnings
warnings.filterwarnings('ignore')

import numpy as nnuumm
import pandas as panndaass
import matplotlib.pyplot as mtpltlb

from sklearn.linear_model import LogisticRegression
from sklearn.svm import LinearSVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier

from sklearn.model_selection import train_test_split, cross_val_score

from sklearn import metrics
from sklearn.metrics import accuracy_score, mean_squared_error, precision_recall_curve, roc_auc_score
from sklearn.model_selection import cross_val_score
```

```python
# Splitting data as train and test
X = dataset1[['id', 'diagnosis', 'radius_mean', 'texture_mean', 'perimeter_mean', 'area_mean', 'smoothness_mean', 'compactness_mean']].astype(str)
y = dataset1['diagnosis'].astype(float)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=0)
```

```python
modelss = {"Logistic_regressionn": LogisticRegression(),
          "K_N_N": KNeighborsClassifier(),
          "Linearr": LinearSVC(),
          "Random_Forest_Cl": RandomForestClassifier()}

def train_scores(models, X_train, X_test, y_train, y_test):
  nnuumm.random.seed(0)
  scoress = {}

  for name, model in models.items():
      model.fit(X_train, y_train)
      scoress[name] = model.score(X_test, y_test)
  return scoress
```

```python
scoress = train_scores(models=modelss, X_train=X_train.values, X_test=X_test.values, y_train=y_train.values, y_test=y_test.values)

scoress
```
